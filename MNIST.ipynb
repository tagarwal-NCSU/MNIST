```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.animation as animation
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import pandas as pd

# Load MNIST database and view the first digit
(train_images_raw, train_labels), (test_images_raw, test_labels) = mnist.load_data()

n_train, h, w = train_images_raw.shape
n_test, h, w = test_images_raw.shape

print("Train Set:", train_images_raw.shape)
print("Test Set:", test_images_raw.shape)

plt.imshow(train_images_raw[0])
plt.title(train_labels[0])
plt.show()
```

```python
# Vectorize the images
train_images = [np.reshape(image, -1) for image in train_images_raw]
test_images = [np.reshape(image, -1) for image in test_images_raw]

# Perform PCA
p = 0.95
pca = PCA(n_components = p, svd_solver = 'full')
train_images_pca = pca.fit_transform(train_images)
test_images_pca = pca.transform(test_images)
```

```python
n = pca.n_components_
pixels = w * h
pct_reduction = (pixels - n) / pixels * 100

print(f"""Each {pixels} pixel image reduced to {n} variables \n{round(pct_reduction, 2)}% percent reduction in data""")
```

```python
# Plot Eigendigits
eigendigits = pca.components_.reshape((n, h, w))

for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(eigendigits[i])
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(train_images_pca, train_labels)
predicted_labels = knn.predict(test_images_pca)
```

```python
frame = 27
plt.imshow(test_images_raw[frame], animated=True)
plt.title(f"Label: {predicted_labels[frame]} Actual: {test_labels[frame]}")
plt.axis('off')
plt.savefig("GoodExample3.png")
plt.show()
```

```python
# Compute accuracy and visualize it on particular labels
acc = accuracy_score(test_labels, predicted_labels)
print(f"Accuracy: {acc*100}%")
cmat = confusion_matrix(test_labels, predicted_labels)
cmat = np.array(cmat)
df = pd.DataFrame(cmat)

x = plt.imshow(cmat, vmax = 40)
plt.colorbar(x, extend = 'max')
plt.ylabel("Actual Labels")
plt.xlabel("Predicted Labels")
plt.savefig('ConfusionMatrix.png')
plt.show()
```

```python
# Bare minimum code to perform analysis
# Lines of code (not counting whitespace or comments): 19
from keras.datasets import mnist
import numpy as np
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load MNIST database
(train_images_raw, train_labels), (test_images_raw, test_labels) = mnist.load_data()

# Get image parameters
n_train, h, w = train_images_raw.shape
n_test, h, w = test_images_raw.shape

# Vectorize the images
train_images = [np.reshape(image, -1) for image in train_images_raw]
test_images = [np.reshape(image, -1) for image in test_images_raw]

# Perform PCA, retaining 95% of variance in data
pca = PCA(n_components = 0.95, svd_solver = 'full')
train_images_pca = pca.fit_transform(train_images)
test_images_pca = pca.transform(test_images)

# Predict labels for test images using kNN with 1 neighbor
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(train_images_pca, train_labels)
predicted_labels = knn.predict(test_images_pca)

# Measure Accuracy
acc = accuracy_score(test_labels, predicted_labels)
print(f"Accuracy: {acc*100}%")
```
